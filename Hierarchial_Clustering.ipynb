#Hierarchial Clustering

Very similar to KMenas

it has two types
1. Agglomerative (Bottom-up approach)
2. Divisive (Top-Down approach)

```
Step 1: Make each data point a single-point-cluster -> that forms N clusters

Step 2: Take the two closest data points and make them one cluster -> that forms N-1

Step 3: Take the two closest clusters and make them one cluster -> that forms N-2

Step 4: Repeat Step 3 until there is only one cluster

FIN
```

closest clusters are closest point from the euclidian distance.

This memory is stored in Dendograms.

###Dendograms

We set a threshold of dissimilarity for clusters in a dendogram.

The line crossed by the threshold line are the number of clusters.

The optimal value of clusters is the vertical distance betweeen the dendogram lines.



##Data Preprocessing

###Importing Libraries

````
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
````

###Importing Dataset

````
df = pd.read_csv('Mall_Customers.csv')
X = df.iloc[:, [3,4]].values
````

##Using the dendogram to find the optimal number of clusters

````
plt.figure(figsize=(12,6))
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
#ward for mininmizing thge variance
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
#ylabel is going to be same no matter what 
````

##Training the heirarchial clustering model on the dataset

````
from sklearn.cluster import AgglomerativeClustering
hac = AgglomerativeClustering(n_clusters = 5, 
                              affinity = 'euclidean',
                              linkage = 'ward')
y_hac = hac.fit_predict(X)
````

````
y_hac
````

````
plt.scatter(X[y_hac==0,0],X[y_hac ==0,1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hac==1,0],X[y_hac ==1,1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hac==2,0],X[y_hac ==2,1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_hac==3,0],X[y_hac ==3,1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_hac==4,0],X[y_hac ==4,1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend() 
````
